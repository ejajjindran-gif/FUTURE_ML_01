# Slide 2 — Introduction

## What Is Strategy?

Strategy guides how businesses differentiate, compete, and create long-term value in changing markets.

## Two Core Strategy Approaches

### Red Ocean Strategy
Competing in an existing, crowded market where companies fight over the same customers.

### Blue Ocean Strategy
Creating a new market space where competition becomes irrelevant through innovation and value creation.

# Slide 3 — Key Differences Between Blue Ocean and Red Ocean Strategy

## Red Ocean Strategy
- Compete in existing markets  
- Focus on outperforming competitors  
- Demand is limited  
- Value–cost trade-off is unavoidable  
- High competition reduces profit margins  

## Blue Ocean Strategy
- Create new, uncontested markets  
- Make competition irrelevant  
- Generate new demand  
- Value innovation: higher value at lower or similar cost  
- Encourages long-term strategic advantage

# Slide 4 — Examples of Strategies

## Blue Ocean Example: Cirque du Soleil
- Blended theatre, dance, and circus  
- Removed costly elements like animals  
- Targeted adult, premium-paying audiences  
- Created a new entertainment category  

## Red Ocean Example: Airline Industry
- Multiple players offering similar services  
- Compete mainly on price and timings  
- Minimal differentiation  
- Low profit margins due to high competition

# Slide 5 — Visual Representation (Conceptual)

## Strategy Comparison Diagram (Text-Based)

Red Ocean:
- High competition  
- Lower value  
- Crowded market  

Blue Ocean:
- Low competition  
- Higher value  
- New market space  

### Concept Summary:
Blue Oceans move upward (value) and rightward (new markets),  
while Red Oceans remain crowded and flat.

# Slide 6 — Conclusion

Business strategy evolves with market conditions.  
Red Oceans focus on competing, while Blue Oceans focus on creating.

## Summary
- Red Oceans: compete for existing demand  
- Blue Oceans: generate new demand  
- Value innovation drives sustainable advantage  

Companies that identify unmet needs and innovate strategically  
are better positioned for long-term growth.
